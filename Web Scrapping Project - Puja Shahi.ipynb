```python
from bs4 import BeautifulSoup as bs
import requests

page=requests.get('https://en.wikipedia.org/wiki/Main_Page', verify = False)

soup=bs(page.text, 'lxml')
titles = soup.find_all(['h1', 'h2','h3'])
for tags in titles:
    print(tags.name + ' -> ' + tags.text.strip())
```

```python
# Display IMDB’s Top rated 100 movies
```

```python
page=requests.get("https://www.imdb.com/list/ls091520106/" , verify = False)
soup=bs(page.content)
mname=soup.find_all('h3',class_='lister-item-header')

mname=soup.find_all('h3',class_='lister-item-header')
mname
movie_name=[]
mv_name=[]
act_mv_name=[]
movie_rating=[]
movie_year=[]

for i in mname:
    movie_name.append(i.text.replace('\n','').replace('(',' ').replace(')','').replace('.',' '))
    

for s in movie_name:
    mv_name.append(s.split(None, 1)[1])    

    
for s in mv_name:
    act_mv_name.append(s[:-5]) 
    
mrating=soup.find_all('div',class_='ipl-rating-star small')
mrating

for i in mrating:
    movie_rating.append(i.text.replace('\n',''))
    
myear=soup.find_all('span',class_='lister-item-year text-muted unbold')
myear

for i in myear:
    movie_year.append(i.text.replace('\n','').replace('(','').replace(')',''))
    
import pandas as pd
df=pd.DataFrame({'Movie':act_mv_name,'Rating':movie_rating,'Year':movie_year})
df    
```

```python
# Display book name, author name, genre and book review of any 5 books
```

```python
page=requests.get("https://bookpage.com/reviews", verify = False)
soup=bs(page.content)
book_name=soup.find_all('h4',class_="italic")

book_names=[]
book_author=[]
book_reviews=[]
book_genre=[]

for i in book_name:
    book_names.append((i.text).replace('\n',''))

author=soup.find_all('p',class_="sans bold")


for i in author:
    book_author.append((i.text).replace('\n',''))
    
review=soup.find_all('p',class_="excerpt")

for i in review:
    book_reviews.append((i.text).replace('\n',''))
    
genre=soup.find_all('p',class_="genre-links hidden-phone")


for i in genre:
    book_genre.append((i.text).replace('\n',''))
    
df=pd.DataFrame({'book':book_names,'author':book_author,'review':book_reviews,'genre':book_genre})
df    
```

```python
# Display 10 ODI teams in men’s cricket along with the records for matches, points and rating
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi", verify = False)
soup=bs(page.content)

top_ODI=soup.find_all('span',class_="u-hide-phablet")
top_ODIs=[]

for i in top_ODI :
    top_ODIs.append((i.text))   
    
while("" in top_ODIs) :
    top_ODIs.remove("")   

match=soup.find_all('td',class_="table-body__cell u-center-text")

match_a=[]
for i in match:
    match_a.append((i.text))

matches = match_a[0::2]

b=soup.find('td',class_="rankings-block__banner--matches")
x=b.text

matches.insert(0,x)
points = match_a[1::2]

c=soup.find('td',class_="rankings-block__banner--points")
y=c.text

points.insert(0,y)

rating=soup.find_all('td',class_="table-body__cell u-text-right rating")
a=soup.find('td',class_="rankings-block__banner--rating u-text-right")
g=a.text.replace("\n",'').replace(" ",'')

rating_a=[]
for i in rating:
    rating_a.append((i.text))
    
rating_a.insert(0,g)   
df=pd.DataFrame({'Team':top_ODIs,'Match':matches,'Point':points,'Rating':rating_a})
df.iloc[0:10,]
```

```python
# Mobile phone details 
```

```python
from urllib.request import urlopen
import re

session = requests.Session()
session.trust_env = False

page=requests.get("https://www.amazon.in/s?k=Mobile+Phones+Under+20000+with+4GB+RAM", verify = False)
soup=bs(page.content)

name=soup.find_all('span',class_="a-size-medium a-color-base a-text-normal")

phone=[]
for i in name:
    phone.append((i.text).replace("\n", ""))  
    
pri=soup.find_all('span',class_="a-price-whole")
price=[]
for i in pri:
    price.append((i.text).replace("\n",""))           

pic_url=soup.find_all('img',class_='s-image')

img=[]

for link in pic_url:
    img.append(link.get('src'))
    
rating=soup.find_all('a',class_='a-popover-trigger a-declarative')

rat=[]
for i in rating:
    rat.append((i.text).replace("\n", ""))   
len(phone),len(rat),len(price),len(img)
    
df=pd.DataFrame({'Product Name':phone,'Price':price,'Imag URL':img,'average rating':rat})
df

```

```python
# Local weather report
```

```python
page=requests.get("https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YMstYWgzZEY", verify = False)
soup=bs(page.content)
period=soup.find_all('p',class_="period-name")

pers=[]
for i in period:
    pers.append(i.text)   

des=soup.find_all('p',class_="short-desc")

desc=[]
for i in des:
    desc.append(i.text)
    
tem=soup.find_all('p',class_="temp")

temp=[]
for i in tem:
    temp.append(i.text) 
    
descrip=soup.find_all('div',class_="col-sm-10 forecast-text")

description=[]
for i in descrip:
    description.append(i.text) 
    
del description[9:]

df=pd.DataFrame({"Period":pers,'Short description':desc,"Temperature":temp,'Description':description})
df

```

```python
# scrape fresher job listings from Internshala
```

```python
page=requests.get("https://internshala.com/fresher-jobs", verify=False)
soup=bs(page.content)

head=soup.find_all('div',class_="heading_4_5 profile")

title=[]
for i in head:
    title.append((i.text).replace("\n", ""))    

comp=soup.find_all('div',class_="heading_6 company_name")
company=[]
for i in comp:
    company.append((i.text).replace("\n", ""))  
    
sal=soup.find_all('div',class_="item_body")

salary=[]

for i in sal:
    salary.append((i.text).replace("\nStarts\xa0", "").replace("\n", "").replace(" ", ""))
salary

del salary[3::3]
salary.pop(0)

app_date=salary[1::2]
sal1=salary[0::2]
    
df=pd.DataFrame({'Job Title':title,'Company':company,'Salary':sal1,'Last apply date':app_date})
df    
```
